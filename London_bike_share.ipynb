```python
#clean up road owner group
import pandas as pd

# Assuming you have a DataFrame called data_explore

# Define the mapping of levels to their replacements
weather_code_description = {
    1: 'Clear',
    2: 'Few clouds',
    3: 'Broken clouds',
    4: 'Cloudy',
    7: 'Light rain',
    10: 'Rain with thunderstorm',
    26: 'Snowfall',
    94: 'Freezing Fog'
}

# Replace the levels in the 'road_owner' column
data['weather_code'] = data['weather_code'].replace(weather_code_description)


# Print the DataFrame with the updated 'road_owner' column
data['weather_code'].value_counts()
    
```

```python
# 1. Impact of Weather on Bike Sharing
plt.figure(figsize=(10, 6))
sns.barplot(x="weather_code", y="cnt", data=df)
plt.title("Impact of Weather on Bike Sharing")
plt.xlabel("Weather ")
plt.ylabel("Bike Share Count")
plt.savefig('weather_bar_plot.png')
plt.show()
```

```python



# Group the data by season and calculate the mean bike sharing counts for each season
seasonal_data = data.groupby('season')['cnt'].mean()

# Define standard colors for each season
colors = ['blue', 'green', 'orange', 'purple']

# Plotting the results
plt.figure(figsize=(10, 6))
seasonal_data.plot(kind='bar', color=colors, alpha=0.7)

# Adding labels and title
plt.xlabel('Season')
plt.ylabel('Average Bike Sharing Counts')
plt.title('Average Bike Sharing Usage Across Seasons')

# Display the plot
plt.show()
```

```python
# 1. Impact of Weather on Bike Sharing
plt.figure(figsize=(10, 6))
sns.barplot(x="weather_code", y="cnt", data=df)
plt.title("Impact of Weather on Bike Sharing")
plt.xlabel("Weather Code")
plt.ylabel("Bike Share Count")
plt.savefig('weather_bar_plot.png')
plt.show()
```

```python
# 2. Seasonal Trends in Bike Sharing
plt.figure(figsize=(10, 6))
sns.boxplot(x="season", y="cnt", data=df)
plt.title("Seasonal Trends in Bike Sharing")
plt.xlabel("Season")
plt.ylabel("Bike Share Count")
plt.savefig('Seasonal trends.png')

plt.show()
```

```python
# 3. Temperature and Bike Sharing
plt.figure(figsize=(10, 6))
sns.scatterplot(x="t1", y="cnt", data=df)
plt.title("Temperature and Bike Sharing")
plt.xlabel("Real Temperature (C)")
plt.ylabel("Bike Share Count")
plt.savefig('Temperature.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="is_holiday", y="cnt", hue="is_weekend", data=df)
plt.title("Impact of Holidays and Weekends on Bike Sharing")
plt.xlabel("Is Holiday")
plt.ylabel("Bike Share Count")
plt.savefig('Holiday.png')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="hum", y="cnt", data=df)
plt.title("Humidity and Bike Sharing")
plt.xlabel("Humidity (%)")
plt.ylabel("Bike Share Count")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['weather_code'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Weather Code Distribution")
plt.show()
```

```python
# 8. Impact of Weather on Weekends
plt.figure(figsize=(12, 8))
sns.lineplot(x="weather_code", y="cnt", hue="is_weekend", data=df)
plt.title("Impact of Weather on Bike Sharing on Weekends vs. Weekdays")
plt.xlabel("Weather Code")
plt.ylabel("Bike Share Count")
plt.savefig('weekend.png')
plt.show()
```

```python
# 9. Temperature "Feels Like" vs. Actual Temperature
plt.figure(figsize=(10, 6))
sns.scatterplot(x="t1", y="t2", data=df)
plt.title("Temperature 'Feels Like' vs. Actual Temperature")
plt.xlabel("Real Temperature (C)")
plt.ylabel("Feels Like Temperature (C)")
plt.show()
```

```python
# 10. Long-Term Trends
plt.figure(figsize=(12, 6))
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.set_index('timestamp')['cnt'].plot()
plt.title("Long-Term Trends in Bike Sharing")
plt.xlabel("Timestamp")
plt.ylabel("Bike Share Count")
plt.show()
```
